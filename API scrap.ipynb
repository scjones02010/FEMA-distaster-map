```python
location='Waynesboro, GA'
category=telecom_cat
```

Get the requests

```python
yelp=[]
offset=None

for i in range(50):
    if offset==None:
        params={'location':location,
           "limit":50,
           'categories':category,
            'offset':offset}
    else:
        params={'location':location,
           "limit":50,
           'categories':category,
            'offset':offset+50}
    url=url
    res=requests.get(url, params=params, headers=header)
    if res.status_code==200:
        yelp_json=res.json()
        yelp.extend(yelp_json['businesses'])
        offset=i
    else:
        print(res.status_code)
        break
    time.sleep(1)
```

```python
item_list=[]
for res in yelp:
    item={}
    item['name']=res['name']
    item['address']=res['location']['display_address']
    item['phone_number']=res['display_phone']
    item['location_latitude']=res['coordinates']['latitude']
    item['location_longitude']=res['coordinates']['longitude']
    item['is_closed']=res['is_closed']
    item['city']=res['location']['city']
    item['zip_code']=res['location']['zip_code']
    item['category']=res['categories'][0]['title']
    item_list.append(item)
    df=pd.DataFrame(item_list)
```
